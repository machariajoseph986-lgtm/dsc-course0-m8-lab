## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:

```python
#ispecting the column names for the aviation data
aviation_df.columns
```

```python
#selecting only the relevant columns for analysis
aviation_cols_df = aviation_df[[
    'Event.Date',
    'Location',
    'Country',
    'Aircraft.Category',
    'Make',
    'Model',
    'Injury.Severity',
    'Aircraft.damage',
    'Number.of.Engines',
    'Engine.Type',
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries'
]]
#checking the first few rows of the subsetted data
df_subset = aviation_cols_df.copy()
df_subset.head()
```

```python
#Filter aircraft types (core step)
filtered_df = df_subset[
    df_subset['Aircraft.Category'].isin(['Airplane', 'Helicopter'])
]
```

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
filtered_df.columns
```

```python
#Ensure injury columns are numeric
injury_cols = [
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries'
]

for col in injury_cols:
    filtered_df.loc[:, col] = pd.to_numeric(filtered_df.loc[:, col], errors='coerce')
```

```python
#Impute missing injury values
for col in injury_cols:
    filtered_df.loc[:, col] = filtered_df.loc[:, col].fillna(0)
```

```python
#We create a composite measure called total_injuries, which aggregates all injury types into a single severity score per accident.
filtered_df.loc[:, 'total_injuries'] = (
    filtered_df['Total.Fatal.Injuries'] +
    filtered_df['Total.Serious.Injuries'] +
    filtered_df['Total.Minor.Injuries']
)
```

```python
#severe injuries
filtered_df = df_subset[
    df_subset['Aircraft.Category'].isin(['Airplane', 'Helicopter'])
].copy()

filtered_df.loc[:, 'severe_injuries'] = (
    filtered_df['Total.Fatal.Injuries'] +
    filtered_df['Total.Serious.Injuries']
)
```

```python
#Estimated passengers
filtered_df['total_injuries'] = (
    filtered_df['Total.Fatal.Injuries'] +
    filtered_df['Total.Serious.Injuries'] +
    filtered_df['Total.Minor.Injuries']
)

filtered_df['estimated_passengers'] = filtered_df['total_injuries'] + 1
```

```python
#fatal injury rate
filtered_df['fatal_injury_rate'] = (
    filtered_df['Total.Fatal.Injuries'] /
    filtered_df['estimated_passengers']
)
```

```python
#serious injury rate
filtered_df['serious_injury_rate'] = (
    filtered_df['Total.Serious.Injuries'] /
    filtered_df['estimated_passengers']
)
```

```python
#validation checks
filtered_df[[
    'total_injuries',
    'severe_injuries',
    'estimated_passengers',
    'fatal_injury_rate',
    'serious_injury_rate'
]].head()
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
#inspect make column
filtered_df['Make'].value_counts().head(20)
```

```python
#clean make column (remove leading/trailing spaces)
filtered_df['Make'] = filtered_df['Make'].fillna("Unknown")
filtered_df['Make'] = filtered_df['Make'].str.upper().str.strip()
```

```python
#filter makes with >=50 occurrences
make_counts = filtered_df['Make'].value_counts()

valid_makes = make_counts[make_counts >= 50].index

filtered_df = filtered_df[filtered_df['Make'].isin(valid_makes)]
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
#Create unique aircraft type ID
filtered_df['aircraft_type'] = (
    filtered_df['Make'] + "_" + filtered_df['Model']
)
```

### Cleaning other columns
- there are other columns containing data that might be related to the outcome of an accident. We list a few here:
- Engine.Type
- Weather.Condition
- Number.of.Engines
- Purpose.of.flight
- Broad.phase.of.flight

Inspect and identify potential cleaning tasks in each of the above columns. Execute those cleaning tasks. 

**Note**: You do not necessarily need to impute or drop NaNs here.

```python
#Clean categorical operational columns
cat_cols = [
    'Engine.Type',
    'Weather.Condition',
    'Purpose.of.flight',
    'Broad.phase.of.flight'
]

for col in cat_cols:
    if col in filtered_df.columns:
        filtered_df[col] = filtered_df[col].fillna("Unknown")
        filtered_df[col] = filtered_df[col].str.strip()
```

```python
# Convert Number.of.Engines to numeric and handle missing values safely
filtered_df['Number.of.Engines'] = pd.to_numeric(
    filtered_df['Number.of.Engines'],
    errors='coerce'
)

# Fill missing engine counts using median (robust to outliers)
filtered_df['Number.of.Engines'] = filtered_df['Number.of.Engines'].fillna(
    filtered_df['Number.of.Engines'].median()
)
```
